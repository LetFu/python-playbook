# Getting Started with Python

This notebook introduces basic Python concepts for beginners.

## Variables and Data Types

Python supports several built-in data types including integers, floats, strings, and booleans.

```python
# Integer
age = 25
print(f"Age: {age}, Type: {type(age)}")

# Float
price = 19.99
print(f"Price: {price}, Type: {type(price)}")

# String
name = "Python Playground"
print(f"Name: {name}, Type: {type(name)}")

# Boolean
is_active = True
print(f"Is Active: {is_active}, Type: {type(is_active)}")
```

## Basic Operations

Python supports arithmetic, comparison, and logical operations.

```python
# Arithmetic operations
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

```python
# Comparison operations
x = 5
y = 10

print(f"{x} == {y}: {x == y}")
print(f"{x} != {y}: {x != y}")
print(f"{x} < {y}: {x < y}")
print(f"{x} > {y}: {x > y}")
print(f"{x} <= {y}: {x <= y}")
print(f"{x} >= {y}: {x >= y}")
```

## String Operations

Python provides powerful string manipulation capabilities.

```python
text = "Hello, Python World!"

# String methods
print(f"Original: {text}")
print(f"Uppercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"Title Case: {text.title()}")
print(f"Replace: {text.replace('Python', 'Notebook')}")
print(f"Split: {text.split(', ')}")
print(f"Length: {len(text)}")
```

```python
# String slicing
text = "Python"

print(f"First character: {text[0]}")
print(f"Last character: {text[-1]}")
print(f"First three: {text[:3]}")
print(f"Last three: {text[-3:]}")
print(f"Reversed: {text[::-1]}")
```

## Control Flow

Python uses if-elif-else statements for conditional execution.

```python
# If-elif-else statement
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")
```

## Loops

Python supports for and while loops for iteration.

```python
# For loop
print("For loop with range:")
for i in range(5):
    print(f"  Iteration {i}")

print("\nFor loop with list:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"  {fruit}")
```

```python
# While loop
print("While loop:")
count = 0
while count < 5:
    print(f"  Count: {count}")
    count += 1
```
